```python
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
```

```python
plt.scatter(x, y, label= "Function: sin(x)")
plt.title('Scatter plot in Matplotlib')
plt.legend()
plt.show
```

```python
plt.figure(figsize=(10,6))

plt.scatter(x, y, label= "y = sin(x)" )

plt.xlabel("X values")
plt.ylabel('Y Values')

plt.title('Scatter Plot in Matplotlib')
plt.legend()
plt.show()
```

```python
np.random.seed(100)
x = np.arange(10)

y = np.random.randn(10)
```

```python

plt.figure(figsize=(10,6))

plt.bar(x, y, label='Sample DATA')
plt.xlabel('X Values - categories')
plt.ylabel('Y Values - Quantities')

plt.title('Bar plot in Matplotlib')
plt.legend()

plt.show
```

```python
np.random.seed(100)
x = 5*np.random.randn(200) + 65

plt.hist(x, bins = 10, edgecolor = 'orange')

plt.xlabel('RETIREMENT AGE')
plt.ylabel('Frequency of values')
plt.title('Histograms in Matplotlib')
plt.show()
```

```python
# Set seed for reproducability
np.random.seed(100)

# Generate a data set of 10000 retirement age values
x = 5*np.random.randn(10000) + 65

#Plot the distogram with hist() function
plt.hist(x, bins=50, edgecolor="black")

plt.xlabel('Retirement Age')
plt.ylabel('Frequency of Values')
plt.title('Histograms in Matplotlib')
plt.show()
```

```python

years = range(1975, 2000)
data = np.linspace(0, 1000, 25) # Create a numpy array of 25 values from 0 - 1000
data
```

```python
#create the plot
fig, ax=plt.subplots()
ax.plot(years, data)

ax.legend(["Sample Data"]);
```

```python
fig, ax= plt.subplots()
ax.plot(years, data)

ax.set_xlabel('X axis lablel')
ax.set_ylabel('Y axis label')

ax.set_title('PLOT TITLE')

ax.legend(["Sample Data"], loc=4);
```

```python
other_fake_data = data - (data % 150) + 50
fig, ax = plt.subplots()

ax.plot(years, data)
ax.plot(years, other_fake_data)

#add the labels
ax.set_xlabel('X Axis label')
ax.set_ylabel('Y Axis Label')

ax.set_title('PLOT TITLE')


ax.legend(['Sample Data', "Other DATA"], loc = 4);

```
